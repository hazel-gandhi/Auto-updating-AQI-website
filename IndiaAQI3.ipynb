```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://cpcb.nic.in/AQI_Bulletin.php"
raw_html = requests.get(url).content
soup_doc = BeautifulSoup(raw_html, 'html.parser')
soup_doc.prettify()
```

```python
pdfs = soup_doc.find('ol').find('li').find('a')['href']
pdfs
```

```python
from datetime import datetime

# URL of the PDF file
url = pdfs

# Get today's date
current_date = datetime.now()

# Format the date for the file name
date_string = current_date.strftime("%Y-%m-%d")

# Create a file name for the PDF
file_name = f"document_{date_string}.pdf"

# Download the PDF
response = requests.get(url, stream=True)

# Save the PDF to a file
with open(file_name, "wb") as file:
    for chunk in response.iter_content(chunk_size=8192):
        file.write(chunk)

print(f"Downloaded and saved the latest file: {file_name}")
```

```python
from img2table.document import PDF

pdf = PDF(file_name)
```

```python
extracted_tables = pdf.extract_tables()

print(f"Found {len(extracted_tables)} tables")
```

```python
extracted_tables[0][0].df
```

```python
df = extracted_tables[0][0].df

# Move the first column into the header
df.columns = df.loc[0]
df = df.drop(index=0)
```

```python
from img2table.document import PDF

pdf = PDF(file_name)

extracted_tables = pdf.extract_tables()

for key, value in extracted_tables.items():
    print(f"{pdf.src}: Page {key} has {len(value)} tables")
```

```python
extracted_tables[1][0].df
```

```python
import pandas as pd

dfs = []

# For every page...
for page in extracted_tables.values():
    # For every table on that page...
    for table in page:
        # Add it to the list
        dfs.append(table.df)


# or be crazy and use a list comprehension
# dfs = [table.df for table in page for page in extracted_tables.values()]

df = pd.concat(dfs, ignore_index=True)
df.head()
```

```python
df.columns = df.loc[0]
df = df.drop(index=0)
df.head()
```
