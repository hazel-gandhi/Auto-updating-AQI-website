```python
from img2table.document import PDF

pdf = PDF("AQI_Bulletin_20241202.pdf")

extracted_tables = pdf.extract_tables()

for key, value in extracted_tables.items():
    print(f"{pdf.src}: Page {key} has {len(value)} tables")
```

```python
extracted_tables[1][0].df
```

```python
import pandas as pd

dfs = []

# For every page...
for page in extracted_tables.values():
    # For every table on that page...
    for table in page:
        # Add it to the list
        dfs.append(table.df)


# or be crazy and use a list comprehension
# dfs = [table.df for table in page for page in extracted_tables.values()]

df = pd.concat(dfs, ignore_index=True)
df.head()
```

```python
df.columns = df.loc[0]
df = df.drop(index=0)
df.head()
```

```python
df.head(40)
```

```python
df_filtered = df[~df['S.No'].str.contains('S.No', case=False, na=False)]
df_filtered.head(100)
```

```python
df = df_filtered[:-7]
df.tail()
```
